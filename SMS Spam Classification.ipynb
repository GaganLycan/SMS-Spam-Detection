```python
# Import the libaries
import pandas as pd
import numpy as np
import string
import re
import nltk
```

```python
df = pd.read_csv(r"E:\\A.I course\\Datasets\SMS spam classification - NLP\\SMSSpamCollection", sep = '\t', names =['labels','messages'])
```

```python
df.head()
```

```python
df.shape
```

```python
# Clean the dataset
```

```python
# Apply stemming
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
```

```python
ps = PorterStemmer()
```

```python
corpus = []
for i in range(0,len(df)):
    review = re.sub('[^a-zA-Z]',' ', df['messages'][i])
    review = review.lower()
    review = review.split()
    review = [ps.stem(word) for word in review if not word in set(stopwords.words('english'))]
    review = ''.join(review)
    corpus.append(review)
```

```python
corpus
```

```python
# Bag of words
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv = CountVectorizer()
```

```python
X = cv.fit_transform(corpus).toarray()
```

```python
X
```

```python
y = pd.get_dummies(df['labels'])
```

```python
y = y.iloc[:,1].values
y
```

```python
# Implement train_test_split
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.33, random_state = 123)
```

```python
# Training the model with Naive bayes classifier
```

```python
from sklearn.naive_bayes import MultinomialNB
```

```python
spam_detector = MultinomialNB().fit(X_train,y_train)
```

```python
y_pred = spam_detector.predict(X_test)
y_pred
```

```python
# Check the accuracy using confusion matrix
```

```python
from sklearn.metrics import confusion_matrix
```

```python
confusion_m = confusion_matrix(y_test,y_pred)
confusion_m
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy = accuracy_score(y_test,y_pred)
accuracy
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_test,y_pred))
```
